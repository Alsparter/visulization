```python
import seaborn as sns
import pandas as pd
```

```python
data=sns.load_dataset('iris') #flower dataset
```

```python
data
```

```python
column = ['sepal_length','sepal_width','petal_length','petal_width','species']
```

```python
data.info()
```

```python
data['species'].value_counts()
```

```python
sns.scatterplot(x=data.sepal_length, y=data.sepal_width)
```

```python
sns.scatterplot(x = data.sepal_length, y = data.petal_length)
```

```python
sns.scatterplot(x=data.petal_width, y=data.petal_length)
```

```python
for i in column:
    sns.displot(x=data[i])
```

```python
sns.relplot(x=data['petal_width'], y= data['petal_length'], data=data)
```

```python
sns.catplot(x='sepal_length', y='species', data=data)
```

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=data)
```

```python
sns.pairplot(data)
```

```python
confuion_matrix= data.corr()
```

```python
sns.heatmap(confuion_matrix,annot=True,linewidths=0.5,)
```
